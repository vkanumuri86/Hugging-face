```python
df=pd.read_csv("medical_cost_dataset.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
X= df[['age', 'bmi', 'children', 'smoker', 'exercise_hours']]
y=df['medical_cost']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30)
```

```python
model =LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred_test = model.predict(X_test)
```

```python
y_pred_test
```

```python
# evaluation metrics : Regression -  R2 (rsquared/adjustedrsquare / root mean sqaure error (rmse) ,meam sqaure error (mse))
```

```python
y_test
```

```python
y_pred_test
```

```python
# calculate errors:
from sklearn.metrics import mean_squared_error , r2_score ,mean_absolute_error
mse = mean_squared_error(y_test,y_pred_test)
mae = mean_absolute_error(y_test,y_pred_test)
r2_score_value = r2_score(y_test,y_pred_test)
print(mse)
print(mae)
print(r2_score_value)
```

```python
rmse = np.sqrt(mse)
print(rmse)
```
